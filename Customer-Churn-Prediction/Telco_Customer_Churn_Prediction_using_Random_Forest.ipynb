```python
# Import the Random Forest machine learning algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load the Telco Customer Churn dataset
df=pd.read_csv('/content/drive/MyDrive/Saved from Chrome/telco.csv')
```

```python
# Display the first 5 rows of the dataset
df.head()
```

```python
# Display dataset information such as columns, data types, and missing values
df.info()
```

```python
# check missing values
df.isnull().sum()
```

```python
# Display dataset columns
df.columns
```

```python
# Remove unnessary columns
df.drop(
    columns=[
        "Customer ID",
        "Churn Score",
        "Customer Status",
        "Churn Category",
        "Churn Reason"
    ],
    inplace=True
)
```

```python
# Fill missing values in the 'Offer' column with 'No Offer'
df['Offer']=df['Offer'].fillna('No Offer')
```

```python
# Fill missing values in the 'Internet type' column with 'No internet service'
df['Internet Type']=df['Internet Type'].fillna('No Internet Service')
```

```python
# convert all catogorical(TEXT) columns into numerical values using Label encoder
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in df.columns:
  if df[column].dtype=='object':
    df[column]=encoder.fit_transform(df[column])
```

```python
# Check the data set columns Data Types
df.dtypes
```

```python
# Create the feature matrix (X) by removing the target column
x = df.drop('Churn Label', axis=1)

# Create the target variable (y) containing the customer churn labels
y = df['Churn Label']
```

```python
# Display the number of rows and columns in the feature dataset
print(x.shape)
```

```python
# Display the number of values in the target variable
print(y.shape)
```

```python
# Display the first 5 rows in feature dataset
x.head()
```

```python
# Display the firt 5 values of target variable
y.head()
```

```python
# split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
#display the shape of traiing and testing datasets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Train the RandonForestClassifier Model
model = RandomForestClassifier(random_state=42)

model.fit(x_train, y_train)

print("Model training completed!")
```

```python
#predict costomer churn using test dataset
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#Calculate accuracy of trained model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:',accuracy)
```

```python
#Display the confusion matrix to evaluate model
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#display the classification report of the model
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```
